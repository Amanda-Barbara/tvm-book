# Python 端部署

加载 TVM Python 环境：

```python
import set_env
```

加载 Python 库：

```python
import tvm
import numpy as np
```

用于部署的常规动态加载方法：

```python
mod_dylib = tvm.runtime.load_module("lib/test_addone_dll.so")
```

从模块中获取函数：

```python
f = mod_dylib["addone"]
```

使用 {func}`tvm.nd.array` 将 numpy ndarray 转换为 tvm ndarray 类型，以便正常调用该函数：

```python
N = 10
x = tvm.nd.array(np.arange(N, dtype=np.float32))
y = tvm.nd.array(np.zeros(N, dtype=np.float32))
```

调用函数：

```python
f(x, y)
```

验证正确性：

```python
np_x = x.numpy()
np_y = y.numpy()
assert np.all([xi + 1 == yi for xi, yi in zip(np_x, np_y)])
```
